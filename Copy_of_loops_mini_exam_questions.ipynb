## 1) Count Up (for + range)
Ask the user for a **positive integer `n`**. Print the numbers **1..n** on **one line** separated by spaces.
- Use a `for` loop with `range`.
- If the input isn’t a positive integer, print a friendly error and stop.


```python
# Write your solution here
# Takes user input
user_input = input("Choose a POSITIVE integer (n):")
# Makes sure user input is an integer
n = int(user_input)
# If input is negative, user will be informed and input
# will not go in for loop
if n <= 0:
  print("Invalid number, n must be a positive integer")
# For every number between 1 and the number the user enters, it will
# be printed on one line seperated by spaces.
for i in range(1, n+1):
  print(i, end=" ")
```

## 2) Sum of Squares (for)
Ask for a **positive integer `n`**. Compute and print **`1^2 + 2^2 + … + n^2`**.
- Use a `for` loop.
- Validate input (print an error and stop if invalid).


```python
# Write your solution here
# Takes user input
user_input = input("Choose a POSITIVE integer (n):")
# Makes sure user input is an integer
n = int(user_input)
# If input is negative, user will be informed and input
# will not go in for loop
if n <= 0:
  print("Invalid number, n must be a positive integer")
  exit()
# Variable that will be used later to keep track of the sum of the
# values of i raised to 2nd power is defined here so a syntax error
# won't occur.
sum = 0
# For every number between 1 and the number the user enters, it will
# be printed on one line seperated by spaces.
for i in range(1, n+1):
  # Keeps track of the sum of values of i raised to the second power
  sum += i ** 2
# result displayed to the user
print(sum)
```

## 3) Min/Max with “done” (while + try/except)
Repeatedly read integers until the user types **`done`**. Ignore invalid entries with a warning.
At the end:
- If at least one valid number was entered, print `min = ___, max = ___`.
- Otherwise print `No numbers entered.`
- Use a `while True:` loop and `break` on `'done'`.


```python
# Write your solution here
# Defines variables min and max so it can be used later on.
min = None
max = None
# Keeps going until user enters "done".
while True:
  # Takes in user input
  user_input = input("Choose an integer:").strip().lower()
  # When user types "done" program will end and if the user entered
  # a number, the results min and max will be printed otherwise
  # it will display "no numbers entered".
  if user_input == "done":
    if type(min) == int and type(max) == int:
      print(f"min = {min} max = {max}")
    else:
      print("No numbers entered")
    break
  # Validates user input, makes sure it is an integer.
  try:
    number = int(user_input)
  # If not an integer, user will be warned and the entry will be ignored.
  except ValueError:
    print("Number must be an integer")
    continue
  # If user input is valid, the integer will be checked to see if it
  # qualifies to be the new minimum or maximum.
  else:
    if min is None or number < min:
      min = number
    if max is None or number > max:
      max = number
```

## 4) Skip & Stop (for + continue/break)
Ask for a **positive integer `n`** (upper bound).
- Loop from **1..n**.
- **Skip** (do not print) numbers that are multiples of **4** using `continue`.
- If you reach **42**, **print 42 and then stop** the loop using `break`.
- Each printed number goes on its own line.
- Validate input; print an error and stop if invalid.


```python
# Write your solution here
# Takes user input
user_input = input("Please choose a positive integer")
# Validates user input, makes sure it is a positive integer,
# otherwise, raises value error and informs user.
try:
  n = int(user_input)
except ValueError:
  print("Invalid input, number must be a positive integer")
  exit()
else:
  if n <= 0:
    print("Invalid number, must be a POSITIVE integer")
    exit()
# I user input is valid, for every value between 1 and n(user input), the
# number will be printed until and including 42. Numbers that are multiples
# of 4 will be skipped.
for i in range(1, n+1):
  if i % 4 == 0:
    continue
  elif i == 42:
    print(i)
    break
  else:
    print(i)
```

## 5) Even Sum with Entry Limit (while + try/except)
Ask for a **target count `k`** (how many entries to read). Then read up to `k` user inputs (one per loop).
- Convert each to `int` with `try/except`.
- Keep a running **sum of even numbers only** (ignore odds and invalid entries).
- After `k` inputs, print `even_sum = ___, valid_evens = ___`.
- If `k` isn’t a positive integer, print an error and stop.


```python
# Write your solution here
# Takes in user input of k
user_input = input(
                  "please enter target count (k)"
                  "(how many entries to read):"
                  )
# if k isn't an integer or if it is a negative integer, the user
# will be informed.
try:
  k = int(user_input)
except ValueError:
  print("Invalid entry, must be an integer")
  exit()
else:
  if k <= 0:
    print("Number must be positive")
    exit()
# Set variables that are used later on so it can be defined
# and a syntax error won't occur.
number_loops = 0
even_sum = 0
valid_evens = 0
# keeps going k amount of times
while number_loops < k:
  # each time loop runs, user asked to enter number
  integer = input("Enter a number:")
  # if the number is odd or if the number is not an integer,
  # entry will be ignored and an entry will be wasted.
  try:
    entry = int(integer)
  except ValueError:
    number_loops += 1
    continue
  else:
    if entry % 2 != 0:
      number_loops += 1
      continue
  # If number is even and valid, value will be added into the even sum,
  # and the number of valid evens will be increased and an entry will be
  # used up.
  number_loops += 1
  valid_evens += 1
  even_sum += entry
# In the end, after the loop is run, results will be displayed.
print(f" even_sum: {even_sum} valid_evens: {valid_evens}")
```
